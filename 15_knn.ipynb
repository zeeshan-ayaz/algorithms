# K_-nearest Neighbour (KNN)

KNN is a ```supervised machine learning``` algorithm that can be sed to solve both ```classification and regression``` problems.

It is a ```non parametric```, lazy learning algorithm. Non-parametric means tath it does not make any assumptios on the underlying data distribution. Lazy learning mean that it does not require any training data points for model generation. All training data used in the testing phase. This makes training faster and testing phase slower cand costlier.



````python
# example of knn classifier on IRIS dataset using sns
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# load the datasset
df = sns.load_dataset("iris")
````

````python
# split the datas into x and y
x = df.drop("species", axis= 1)
y = df['species']
````

````python
# load and fit the knn classifier on the data
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x,y)
````

````python
# fit the model 
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
````
