```python
# load the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# drop deck colm
df.drop('deck', axis=1,  inplace=True)
```

```python
df.info()
```

```python
# impute missing value of age and fare using median
imputer = SimpleImputer(strategy='mean') 
df[['age', 'fare']] = imputer.fit_transform([['age', 'fare']])

# impute missing value of embark town using mode
imputers = SimpleImputer(strategy='most_frequent')
df[['embark_town', 'embarked']] = imputers.fit_transform([['embark_twon', 'embarked']])
```
